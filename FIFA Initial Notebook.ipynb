Below have created a function which will subset the Fifa dataframes. The function will also filter the records to only keep th players which belong to the leagues Spain Primera Division, German 1. Bundesliga, French Ligue 1, English Premier League, Italian Serie A

```python
subset = ['short_name', 'long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'club_name', 'league_name', 'overall', 
          'potential', 'value_eur', 'wage_eur', 'player_positions', 'preferred_foot', 'release_clause_eur', 
          'nation_position', 'nation_jersey_number', 'previous_edition_overall']
```

```python
leagues = ['Spain Primera Division', 'German 1.', 'Bundesliga', 'French Ligue 1', 'English Premier League', 'Italian Serie A']
```

Creating a function which creates a new column 'Name_Formatted'. It creates an entry in this column by splitting the strings in the name column. If there is only one string, then that string will be the new entry. Otherwise, we take the first letter of the first string, put a '.' and then put the last string

```python
def name_format(name):
    names = name.split()
    if len(names) == 1:
        return name
        
    fn = names[0]
    ln = names[-1]
    
    new_name = f'{fn[0]}. {ln}'
    return new_name
```

```python
Season_df['Name_Formatted'] = Season_df['Player Name'].apply(name_format)
```
