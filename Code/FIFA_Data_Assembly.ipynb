## Fuzzy Match and Name Matching

We use the library fuzzywuzzy to match names across our dataset. Between the fifa dataset and the real dataset, the same player can have different lengths of names and this makes it difficult to merge the two datasets.

We use the Fuzzy match approach to find similar names in the dataset and then standardise these names between the two datasets to enable merging.

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
import recordlinkage
```

```python
df1.index = np.arange(len(df1))
```

```python
df2.index = np.arange(len(df2))
```

```python
pd.DataFrame(df1.columns)
```

```python
df2[df2['Team']=='BAR']
```

To reduce the time taken to scan through all the names to find similar names, we use a blocking method to search for names only by season and team. so For example, if we pick any player from a team for a particular season in the FIFA dataset, we will search for similar names in the real dataset after filtering for the same team and season.

This helps us to avoid scanning the entire dataset to search for similar names and also provides higher confidence in the outputs of the fuzzy match approach.

```python
indexer = recordlinkage.Index()
```

```python
%%time
indexer.block(['Season', 'Team'])
multi_index = indexer.index(df1, df2)
print(multi_index)
```

```python
%%time
threshold= 95
import collections
dict1=collections.defaultdict(list)


for i in multi_index:
    if (fuzz.partial_ratio(df1.iloc[i[0], 19], df2.iloc[i[1], 2]))>threshold:
        dict1[df1.iloc[i[0], 19]].append(df2.iloc[i[1], 2])
print (dict1)
```

```python
df3=pd.DataFrame(dict1.items())
df3
```

The above dictionary is the final dictionary which is used to match and standardise names between the FIFA and Real datasets

```python
df1.to_csv('FIFA_new.csv', index=False, header=True)
df2.to_csv('Real_new.csv', index=False, header=True)
df3.to_csv('Name_matching1.csv', index=False, header=True)
```
